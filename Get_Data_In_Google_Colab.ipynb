#Q1) To read a CSV file in Google Colab

```python
import pandas as pd
df = pd.read_csv("Subject1.csv")
print(df.head())
```

```python
from google.colab import files
import pandas as pd
uploaded = files.upload()
csv_file = list(uploaded.keys())[0]
df = pd.read_csv(csv_file)
print(df.head())
```

#Q2) To read data from an Excel file in Google Colab

```python
import pandas as pd
df = pd.read_excel("demo.xlsx")
print(df.head(2))
```

```python
from google.colab import files
import pandas as pd
uploaded = files.upload()
excel_file = list(uploaded.keys())[0]
df = pd.read_excel(excel_file)
print(df.head(2))
```
